## import library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## membaca gambar

```python
image = cv2.imread('asset/nama no shadow.jpg')
b, g, r = cv2.split(image)
gambar_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## mencari ambang batas pertama

```python
threshold_r = 100
threshold_g = 100
threshold_b = 100

_, r_mask = cv2.threshold(r, threshold_r, 255, cv2.THRESH_BINARY_INV)
_, g_mask = cv2.threshold(g, threshold_g, 255, cv2.THRESH_BINARY_INV)
_, b_mask = cv2.threshold(b, threshold_b, 255, cv2.THRESH_BINARY_INV)

rgb_mask1 = cv2.bitwise_or(cv2.bitwise_and(r_mask, g_mask), b_mask)

none_mask = np.zeros_like(r_mask)

plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 1)
plt.imshow(gambar_rgb)
plt.title("Gambar Asli")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(none_mask, cmap='gray')
plt.title("NONE")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(rgb_mask1, cmap='gray')
plt.title("RED")
plt.axis('off')

plt.tight_layout()
plt.show()
```

## mencari ambang batas kedua

```python
threshold_r = 147
threshold_g = 147
threshold_b = 147

_, r_mask = cv2.threshold(r, threshold_r, 255, cv2.THRESH_BINARY_INV)
_, g_mask = cv2.threshold(g, threshold_g, 255, cv2.THRESH_BINARY_INV)
_, b_mask = cv2.threshold(b, threshold_b, 255, cv2.THRESH_BINARY_INV)

rgb_mask2 = cv2.bitwise_or(cv2.bitwise_and(r_mask, g_mask), b_mask)

none_mask = np.zeros_like(r_mask)

plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 1)
plt.imshow(gambar_rgb)
plt.title("Gambar Asli")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(none_mask, cmap='gray')
plt.title("NONE")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(rgb_mask2, cmap='gray')
plt.title("RED-GREEN")
plt.axis('off')

plt.tight_layout()
plt.show()
```
