## import library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## membaca gambar

```python
img = cv2.imread('asset/diri.jpg')
```

## citra asli

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 12))
plt.imshow(img_rgb)
plt.title("Gambar Asli (RGB)")
plt.axis('off')
plt.show()
```

## citra grayscale


```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12, 12))
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')
plt.show()
```

## citra cahaya

```python
bright = cv2.convertScaleAbs(gray, alpha=1.0, beta=100)  

plt.figure(figsize=(12, 12))
plt.imshow(bright, cmap='gray')
plt.title("Brightness Enhanced")
plt.axis('off')
plt.show()
```

# citra kontras

```python
contrast = cv2.convertScaleAbs(gray, alpha=3.0, beta=0)  
plt.figure(figsize=(12, 12))
plt.imshow(contrast, cmap='gray')
plt.title("Contrast Enhanced")
plt.axis('off')
plt.show()
```

## citra kontras & cahaya

```python
bright_contrast = cv2.convertScaleAbs(gray, alpha=3.0, beta=100)

plt.figure(figsize=(12, 12))
plt.imshow(bright_contrast, cmap='gray')
plt.title("Brightness + Contrast Enhanced")
plt.axis('off')
plt.show()
```
