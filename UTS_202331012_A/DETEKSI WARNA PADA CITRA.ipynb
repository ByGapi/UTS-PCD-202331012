## import library


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## membaca gambar


```python
gambar = cv2.imread('asset/nama.jpg')
```

## citra asli

```python
gambar_rgb = cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB)

colors = ('r', 'g', 'b')
channel_ids = (0, 1, 2)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(gambar_rgb)
plt.title("Gambar Asli")
plt.axis('off')

plt.subplot(1, 2, 2)
for channel_id, color in zip(channel_ids, colors):
    histogram = cv2.calcHist([gambar], [channel_id], None, [256], [0, 256])
    plt.plot(histogram, color=color)
plt.title("Histogram RGB Gambar Asli")
plt.xlabel("Intensitas")
plt.ylabel("Jumlah Piksel")

plt.tight_layout()
plt.show()
```

## deteksi merah

```python
b, g, r = cv2.split(gambar)

merah_only = cv2.merge([np.zeros_like(b), np.zeros_like(g), r])

merah_gray = cv2.cvtColor(merah_only, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(merah_gray, cmap='gray')
plt.title("Deteksi Warna Merah")
plt.axis('off')

plt.subplot(1, 2, 2)
hist_merah = cv2.calcHist([merah_gray], [0], None, [256], [0, 256])
plt.plot(hist_merah, color='red')
plt.title("Histogram Deteksi Warna Merah")
plt.xlabel("Intensitas")
plt.ylabel("Jumlah Piksel")

plt.tight_layout()
plt.show()
```

## deteksi hijau

```python
b, g, r = cv2.split(gambar)

hijau_only = cv2.merge([np.zeros_like(b), g, np.zeros_like(r)])

hijau_gray = cv2.cvtColor(hijau_only, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(hijau_gray, cmap='gray')
plt.title("Deteksi Warna Hijau")
plt.axis('off')

plt.subplot(1, 2, 2)
hist_hijau = cv2.calcHist([hijau_gray], [0], None, [256], [0, 256])
plt.plot(hist_hijau, color='green')
plt.title("Histogram Deteksi Warna Hijau")
plt.xlabel("Intensitas")
plt.ylabel("Jumlah Piksel")

plt.tight_layout()
plt.show()
```

## deteksi biru


```python
b, g, r = cv2.split(gambar)

biru_only = cv2.merge([b, np.zeros_like(g), np.zeros_like(r)])

biru_gray = cv2.cvtColor(biru_only, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(biru_gray, cmap='gray')
plt.title("Deteksi Warna Biru")
plt.axis('off')

plt.subplot(1, 2, 2)
hist_biru = cv2.calcHist([biru_gray], [0], None, [256], [0, 256])
plt.plot(hist_biru, color='blue')
plt.title("Histogram Deteksi Warna Biru")
plt.xlabel("Intensitas")
plt.ylabel("Jumlah Piksel")

plt.tight_layout()
plt.show()
```
